## 1.A.4 Do these ARAs and RRAs increase or decrease with initial wealth? 

```python
x = np.linspace(1,10,10)
power_ara = np.empty(len(x))
power_rra = np.empty(len(x))
quad_ara = np.empty(len(x))
quad_rra = np.empty(len(x))
exp_ara = np.empty(len(x))
exp_rra = np.empty(len(x))


for cons in range(1,10):
    y_func = sy.lambdify((sigma,c), u_power_object.ara())
    power_ara[cons] = y_func(.7, cons+2)
    
    y_func = sy.lambdify((sigma), u_power_object.rra())
    power_ara[cons] = y_func(.7)
    
    y_func = sy.lambdify((b,c), u_quad_object.ara())
    quad_ara[cons] = y_func(1,cons+2)
    
    y_func = sy.lambdify((b,c), u_quad_object.rra())
    quad_rra[cons] = y_func(1, cons+2)
    
    y_func = sy.lambdify((a), u_exp_object.ara())
    exp_ara[cons] = y_func(1)
    
    y_func = sy.lambdify((a,c),  u_exp_object.rra())
    exp_rra[cons] = y_func(1,cons+2)
    
plt.plot(x,power_ara, label='Power ara')
plt.plot(x, power_rra, label= 'Power rra')
plt.plot(x, quad_ara, label= 'Quad ara')
plt.plot(x, quad_rra, label= 'Quad rra')
plt.plot(x, exp_ara, label= 'Exp ara')
plt.plot(x, exp_rra, label= 'Exp rra')


plt.legend();
#plt.plot(x, quad_rra, label= 'Power rra')
```
